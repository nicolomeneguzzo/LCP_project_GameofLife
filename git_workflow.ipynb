# Workflow di GitHub

Inizialmente creiamo una direcotry nella quale lavoreremo e ci spostiamo dentro di essa.

## Creare una repository

Come prima cosa bisogna andare su github e cliccare su "new". Ora si possono compilare i campi:
1. Repository name
2. Description (facoltativa)
3. Seleziona Public/Private
4. Add README
5. Add .gitignore
6. Add license



### Description
La descrizione è facolatativa ma utile. Di solito si tratta di una breve frase con una rapida spiegazione di cosa la repository conterrà. La descrizione comparirà sotto il nome della repo.

### Scelta Public/Private
In una repo pubblica, chiunque può vederla senza bisogno di inviti. Questo permette collaborazioni, una maggiore visibilità e feedback.\
In una repo privata invece, la repo non è accessibile senza permessi.Questo garantisce sicurezza e controllo totale su chi può accedere al contenuto.

### README
Il README è il file principale di una repository che serve a spiegare di cosa tratta il progetto. Serve per permettere a chiunque apra la repo, incluso te stesso in futuro, a capire rapidamente il progetto. In pratica, è come un manuale di istruzioni iniziale per il progetto. Un README ben fatto può includere:Titolo del progetto, Descrizione breve (a cosa serve il progetto), Installazione / setup (come avviare il progetto), Esempi d’uso, Tecnologie usate (Python, Node.js, React ecc.), Autori o contributori, Licenza (opzionale).\
Il README è modificabile in ogni momento. è sufficiente creare un file README.md nella repo locale e fare commit.

### .gitignore
Il file .gitignore è un semplice file di testo che dice a Git: “Ignora questi file o cartelle, non aggiungerli al repository.” In altre parole, Git non traccerà i file elencati in .gitignore. Questo permette di non far finire tutti i file nella directory su github.\
Per crearla è sufficiente creare un file chiamato .gitignore nella cartella principale della repo:

`nano .gitignore`

successivamente dentro il file vengono scritti i documenti che vogliamo ignorare.

## Creare il branch develop

`git checkout -b develop`

git checkout serve a cambiare branch. -b develop significa “crea un nuovo branch e passa subito a esso”.\
Dopo il comando, puoi vedere su quale branch sei:

`git branch`

Ora si può mettere il develop su github:

`git push -u origin develop`

git push serve a inviare i commit locali al repository remoto, in pratica, aggiorna la repo online con le modifiche che hai fatto in locale.\
origin è il nome del repository remoto predefinito quando cloni una repo.\
develop specifica il branch da inviare al remoto.\
-u (o --set-upstream) imposta il branch remoto come “upstream” del branch locale. Significa che da quel momento puoi fare solo git push o git pull senza dover scrivere origin develop ogni volta.

`git branch -a`

mostra tutti i branch, sia locali che remoti

## Workflow standard
### Per iniziare

Prima di tutto bisogna essere sicuri di essere su develop. Nel caso non lo fossimo bisogna spostarsi sul develop e poi aggiornarlo.

`git branch`

`git checkout develop` oppure `git switch develop`

`git pull`

Più volte si aggiorna il develop e meno si rischiano conflitti\
Ora creiamo un nuovo branch nel quale lavorare. Io lo chiamo "personal"

`git checkout -b personal` oppure `git switch -c personal`

è fondamentale che ognuno lavori sul proprio branch personale altrimenti si creano bug.\
ATTENZIONE: anche solo per aprire un progetto jupyter bisogna essere sul branch personale, perchè anche solo aprire un file jupyter crea modifiche che devono essere gestite.

Se il lavoro svolto non ti soddisfa, puoi eliminare il branch: `git branch -d personal` e successivamente ricrearlo

### git add & git commit

Dopo aver lavorato sul proprio branch personale bisogna fare:

(con `git status` puoi vedere quali file sono stati modificati)

`git add file_che_voglio_committare` oppure `git add -A` se vuoi committare tutti i file della repo

`git commit -m "commento"`

ATTENZIONE: se si vuole passare da personale a develop ma il lavoro su personal non è finito bisogna fare comunque add e commit altrimenti nel cercare di passare da uno all'altro andranno in conflitto. se invece si fa prima add e commit il conflitto non capita

Vediamo cosa succede con questo passaggio. Un commit è una fotografia salvata del progetto in un determinato momento. È un punto nella storia del progetto. Il commit serve per salvare le modifiche. Tuttavia queste modifiche non vengono salvate direttamente. git ha una zona intermedia chiamata Staging Area alla quale si accede con `add`. il `git commit` aggiornerà il progetto salvando tutto ciò che hai inserito nella Staging Area. In questo modo posso decidere di mettere nella Staging Area diversi documenti ai quali sto lavorando, quando faccio il commit salvo le modifiche. Se si fa il commit senza fare prima add il commit sarà vuoto. La staging area contiene solo le versioni dei file che voglio salvare con il prossimo commit.


Immaginiamo di aver lavorato sul file chiamato "file.ipynb". Posso fare `git add file.pynb` e il file verrà inserito nella staging area, pronto per essere committato. Poi però mi accorgo che devo modificarlo ancora. Una volta modificato devo rifare `git add file.ipynb`. In questo modo la versione del file nella staging area viene sovrascritta con la nuova versione e posso fare `git commit`. 

### Merging

Dopo aver fatto il commit abbiamo salvato le modifiche sul branch "personal". Prima di fare merge bisogna aggiornare nuovamente il develop. Questo evita conflitti nel caso in cui qualcuno abbia pushato modifiche mentre lavoravi.

`git switch develop`

`git pull`

In questo modo il tuo develop locale viene aggiornato alla versione più recente. Se altri hanno modificato i file, ora queste modifiche sono presenti nel develop locale.\
Ora si ritorna al branch "personal" e si fa merge.

`git switch personal`

`git merge develop`

Questo primo merge serve per inserire le eventuali modifiche del develop all'interno del tuo branch personal. Ora il tuo branch ha sia tutte le eventuali modifiche fatte da altri che le tue precedenti modifiche. In questo modo se ci sono conflitti possono essere risolti direttamente dentro il branch personale.

Ora si ritorna sul develop e si fa merge

`git switch develop`

`git merge personal`

In questo secondo merge abbiamo salvato le nostre modifiche sul develop locale e possiamo procedere con il push e l'eliminazione del branch personal.

`git push`

`git branch -d personal`

## Conflitti

Un conflitto (merge conflict) si verifica quando Git non riesce a unire automaticamente due versioni di uno stesso file, perché le stesse righe sono state modificate in due branch diversi e Git non sa quale versione mantenere. Succede solo al momento del merge di due branch. La risoluzione richiede l’intervento manuale dello sviluppatore, che deve scegliere quale versione mantenere o combinare entrambe le modifiche. Se due persone lavorano sullo stesso file ma modificano righe diverse, Git riesce a unire automaticamente le modifiche senza problemi. I conflitti sorgono solo quando le stesse righe vengono modificate in branch diversi.

### Esempio

Immaginiamo che Alice e Bob stiano lavorando sullo stesso file chiamato "saluto.txt". Il file nel branch develop remoto (su github) contiene la parola "ciao". Sia Alice che Bob hanno nel loro develop locale la parola "ciao". Entrambi stanno lavorando sul loro branch personale locale. Alice cambia la parola in "buongiorno" in "personal_alice" mentre Bob la cambia in "buonasera" in "personal_bob". Al momento non ci sono conflitti, ognuno sta lavorando sul proprio progetto. Ora immaginiamo che alice svolga la procedura di merging.\
`git switch develop`

`git pull` (In questo caso notiamo che il develop locale di alice e remoto sono uguali e quindi non serve fare merge delle modifiche nel branch personale)

`git merge personal_alice`

`git push`

Il merge in questo caso è risucito senza conflitti.\
\
Ora Bob prova a fare la stessa cosa.\
Prima esegue:\
`git switch develop`

`git pull` 

ora il develop locale di Bob contiene "buongiorno". Se ora Bob prova a integrare develop nel suo branch con

`git switch personal_bob` (personal_bob contiene "buonasera" mentre develop buongiorno)

`git merge develop` 
Ora abbiamo conflitto!


Per risolvere il conflitto bob deve aprire il file in cui c'è conflitto, ovvero saluto.txt da personal_bob

`nano saluto.txt`

all'interno del file comparirà questo:

```
<<<<<<< HEAD
buonasera
=======
buongiorno
>>>>>>> develop
```

Questo ci dice che HEAD ovvero il branch personal_bob, contiene la parola buonasera mentre develop contiene buongiorno. Dobbiamo decidere cosa tenere cancellado la parola che non vogliamo e i maggiori, minori e uguali. Immaginiamo di tenere "buonasera"\
Ora si fa `git add saluto.txt` e `git commit`. Si aprirà un editor e dobbiamo solo salvare.
Una volta risolto il conflitto hai 2 possibilità. O Bob continua a lavorare sul branch personal_bob con conflitto risolto (perchè ora il suo branch personale contiene sia il develop aggiornato che le sue modifiche). oppure può pushare le sue modifiche nel remoto (ora il develop locale contiene ancora la parola buongiorno). allora:\
`git switch develop`

`git merge personal_bob`

`git push`

### Altre cose utili

`git restore file` questo comando permette di ripristinare un file allo stato dell'ultimo commit nel branch corrente e scartare le modifiche locali
